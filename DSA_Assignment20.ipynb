```python
#Q1 - 
'''  
Given a binary tree, your task is to find subtree with maximum sum in tree.

Input for this question is :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7

Output of this question should be: 28
'''

#Answer:
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def find_max_subtree_sum(root):
    max_sum = float('-inf')  # Initialize maximum sum as negative infinity

    def dfs(node):
        nonlocal max_sum

        if node is None:
            return 0

        # Calculate the sum of the subtree rooted at the current node
        subtree_sum = node.value + dfs(node.left) + dfs(node.right)

        # Update the maximum sum if the current subtree's sum is greater
        max_sum = max(max_sum, subtree_sum)

        return subtree_sum

    dfs(root)  # Start DFS traversal from the root node

    return max_sum


# Create the binary tree from the given input
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Find the subtree with the maximum sum
max_subtree_sum = find_max_subtree_sum(root)

print("Maximum subtree sum:", max_subtree_sum)
```

```python
#Q2 - 
''' 
Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input of this question is: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output of this question should be: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10
'''

#Answer:
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def construct_bst(level_order):
    if not level_order:
        return None

    # Create the root node
    root = Node(level_order[0])

    # Traverse through each element in the level order traversal array
    for i in range(1, len(level_order)):
        insert_node(root, level_order[i])

    return root


def insert_node(node, value):
    if value < node.value:
        # Insert in the left subtree
        if node.left is None:
            node.left = Node(value)
        else:
            insert_node(node.left, value)
    else:
        # Insert in the right subtree
        if node.right is None:
            node.right = Node(value)
        else:
            insert_node(node.right, value)


def print_bst_inorder(root):
    if root:
        print_bst_inorder(root.left)
        print(root.value, end=" ")
        print_bst_inorder(root.right)


# Given level order traversal array
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Construct the BST
bst_root = construct_bst(level_order)

# Print the BST inorder
print("BST:")
print_bst_inorder(bst_root)
```

```python
#Q3 - 
'''  
Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input of this question is : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output of this question should be : No

For the given arr[], the Binary Search Tree is:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10
'''

#Answer:
def can_represent_bst_level_order(arr):
    n = len(arr)

    if n == 0:
        return "No"  # Empty array cannot represent a BST

    # Initialize lower_bound and upper_bound with minimum and maximum values
    lower_bound = float('-inf')
    upper_bound = float('inf')

    i = 0
    while i < n:
        curr = arr[i]

        if curr < lower_bound or curr > upper_bound:
            return "No"  # Current element is not within the valid range

        # Update the valid range for the next nodes
        if i + 1 < n and arr[i + 1] < curr:
            upper_bound = curr
        if i + 2 < n and arr[i + 2] > curr:
            lower_bound = curr

        i += 2

    return "Yes"  # The array can represent a valid BST


# Given level order traversal array
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Check if the array can represent a BST
result = can_represent_bst_level_order(arr)

print(result)
```
